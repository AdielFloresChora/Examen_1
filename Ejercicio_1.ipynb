```python
import random
Persona_1 = random.uniform(0, 30)
Persona_2 = random.uniform(0, 30)
```

Determine la distribución de probabilidad<br>
Como ambas personas tienen la misma probabilidad de llegar al lugar de la cita en cualquier momento dentro del intervalo de tiempo [0, 30], se puede asumir que las llegadas siguen una distribución uniforme en ese intervalo.

Modele las variables aleatorias<br>
El modelo del sistema es el tiempo que una persona espera a la otra. El objetivo es determinar la probabilidad de que el tiempo de espera sea a lo más 10 minutos.

```python
Diferencia = abs(Persona_1-Persona_2)
```

Diseñe y repita el experimento n veces<br>
Se genera n veces las variables aleatorias entre el intervalo [0, 30].<br>
Si la diferencia entre las variables es menor o igual a 10, se suma en 1 la variable "prob".<br>
Se obtiene la probabilidad dividiendo "prob" entre n.

```python
prob = 0
n = 1000
for i in range(n):
    Persona_1 = random.uniform(0, 30)
    Persona_2 = random.uniform(0, 30)
    Diferencia = abs(Persona_1-Persona_2)
    if Diferencia <= 10:
        prob += 1
prob/n
```

Finalmente,

```python
import random
import matplotlib.pyplot as plt

def Llegada(n):
    Prob = 0
    L = []
    for i in range(n):
        Persona_1 = random.uniform(0, 30)
        Persona_2 = random.uniform(0, 30)
        Diferencia = abs(Persona_1-Persona_2)
        if Diferencia <= 10:
            Prob += 1
        L.append(Prob/(i+1))
    plt.plot(L)
    plt.xlabel("Número de simulaciones")
    plt.ylabel("Probabilidad")
    plt.show()
    print("Se tiene que la probabilidad de espera sea como máximo 10 es de",Prob/n*100,"%")
    return 

Llegada(10000)
```

Ahora se resuelve de manera analítica.

Se tiene que si ambas personas tienen la misma probabilidad de llegar al lugar de la cita en el intervalo de tiempo [0, 30], entonces el tiempo de llegada de ambos sigue una distribución uniforme con parámetros a=0 y b=30.<br>
Sea X el tiempo de llegada de la primera persona y Y el tiempo de llegada de la segunda persona. Así, el tiempo de espera de una persona a otra sea como máximo 10 minutos es de P(X-Y)<= 10.<br>
Para obtener esta probabilidad se debe obtener la función de densidad conjunta de X e Y y posteriormente integrar sobre la región donde |X-Y|<=10<br>
La función de densidad conjunta f(x, y) = (1/30)*(1/30) = 1/900.<br>
Se tiene la doble integral  $$\int\limits_{0}^{30}\int\limits_{x}^{30} \frac{1}{900} dy dx = \dfrac{1}{2}$$.


Así, la probabilidad de que el tiempo de espera sea como máximo 10 minutos es de $\dfrac{1}{2}$
